1 -Utilize o Pandas para carregar o conjunto de dados "Breast Cancer" do UCI Machine Learning Repository em um dataframe, e imprima as 10 primeiras linhas.

```python

# Inicializção das bibliotecas
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer/breast-cancer.data"

# Lê e carrega o dataset na memória
df = pd.read_csv(url)

df.head(10)
```

2 - Substitua o cabeçalho pelos nomes no arquivo breast-câncer.names

```python

# Define o nome das colunas
header = ['Class', 'age', 'menopause', 'tumor-size', 'inv-nodes','node-caps','deg-malig','breast','breast-quad','irradiat']
# Lê e carrega o arquivo para a memória
df = pd.read_csv(url, header=None, names=header)

df.head(10)
```

3 - Use o método info() do Pandas para obter informações sobre o dataframe, como o número de entradas, o tipo de dados de cada coluna, entre outras informações.

```python
df.info()
```

4 - Criar um subdataframe contendo apenas o conteúdo das colunas Class, age, menopause e tumor-size

```python
# Criar um subdataframe
df['Class'] = df['Class'].map({'no-recurrence-events': 0, 'recurrence-events': 1})
subdf = df[['Class', 'age', 'menopause', 'tumor-size']]

subdf.head(10)
```
